```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import datetime as date
now = date.datetime.now()
```

### Plotting the Data

```python
weatherdf = pd.read_csv("../Assets/weather_data.csv")
```

#### Latitude vs. Temperature Plot

```python
plt.rc('figure', figsize=(14,10))
plt.rcParams.update({'font.size':35})


fig, ax = plt.subplots()
ax.set_title(f'Latitutde vs Temperature \n ({now.month}/{now.day}/{now.year})\n')
weatherdf.plot(kind='scatter', 
               x='Latitude', 
               y='Temperature', 
               c='Longitude', 
               cmap='Reds', 
               grid=True, 
               ax=ax)
ax.set_axisbelow(True)
plt.savefig('../Assets/MaxTemp.png')
```

#### Latitude vs. Humidity Plot

```python
fig, ax = plt.subplots()
ax.set_title(f'Latitutde vs Humidity\n{now.month}/{now.day}/{now.year}\n')
weatherdf.plot(kind='scatter', 
               x='Latitude', 
               y='Humidity', 
               c='Longitude', 
               cmap='Reds', 
               ax=ax,
               grid=True) 
ax.set_axisbelow(True)

plt.savefig('../Assets/Humidity.png')
```

#### Latitude vs. Cloudiness Plot
